```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

# Load data
df = pd.read_csv("/content/data (5) (1) (1) (1) (1)(internship-data-1).csv", parse_dates=['time'])
df.set_index('time', inplace=True)

# Select univariate time series (e.g., Cyclone_Material_Temp)
series = pd.to_numeric(df['Cyclone_Material_Temp'], errors='coerce').dropna()

# Optional: downsample to hourly to reduce size (comment out if unnecessary)
series = series.resample('H').mean().dropna()

# Train ARIMA model
model = ARIMA(series, order=(5, 1, 2))  # you can tune p,d,q
model_fit = model.fit()

# Predict (in-sample forecast)
forecast = model_fit.predict(start=series.index[1], end=series.index[-1], typ='levels')
residuals = series[1:] - forecast

# Anomaly threshold: 3 standard deviations of residuals
threshold = 3 * residuals.std()
anomalies = residuals[np.abs(residuals) > threshold]

# Plot
plt.figure(figsize=(15, 6))
plt.plot(series, label='Actual')
plt.plot(forecast, label='ARIMA Forecast', alpha=0.7)
plt.scatter(anomalies.index, series[anomalies.index], color='red', label='Anomalies')
plt.title('ARIMA Anomaly Detection on Cyclone_Material_Temp')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from scipy.stats import kurtosis, skew
import numpy as np

# --- Step 1: Load and prepare data ---
df = pd.read_csv("/content/cyclonedata.csv", parse_dates=['time'])
df.set_index('time', inplace=True)

# Ensure sensor columns are numeric
sensor_cols = [
    'Cyclone_Inlet_Gas_Temp',
    'Cyclone_Gas_Outlet_Temp',
    'Cyclone_Outlet_Gas_draft',
    'Cyclone_cone_draft',
    'Cyclone_Inlet_Draft',
    'Cyclone_Material_Temp'
]
df[sensor_cols] = df[sensor_cols].apply(pd.to_numeric, errors='coerce')
df.dropna(subset=sensor_cols, inplace=True)

# --- Step 2: Row-wise statistics ---
df['row_mean'] = df[sensor_cols].mean(axis=1)
df['row_std'] = df[sensor_cols].std(axis=1)
df['row_max'] = df[sensor_cols].max(axis=1)
df['row_min'] = df[sensor_cols].min(axis=1)
df['row_skew'] = df[sensor_cols].apply(lambda x: skew(x, nan_policy='omit'), axis=1)
df['row_kurtosis'] = df[sensor_cols].apply(lambda x: kurtosis(x, nan_policy='omit'), axis=1)

# --- Step 3: ARIMA anomaly detection function ---
def detect_arima_anomalies(series, threshold=3):
    series = series.dropna()
    model = ARIMA(series, order=(5, 1, 0))
    model_fit = model.fit()
    forecast = model_fit.predict(start=series.index[1], end=series.index[-1], typ='levels')
    residuals = series[1:] - forecast
    std = residuals.std()
    anomalies = abs(residuals) > threshold * std
    return anomalies.reindex(series.index, fill_value=False)

# --- Step 4: Apply ARIMA on all selected columns ---
all_features = sensor_cols + ['row_mean', 'row_std', 'row_max', 'row_min', 'row_skew', 'row_kurtosis']
anomaly_flags = pd.DataFrame(index=df.index)

for col in all_features:
    print(f"Processing: {col}")
    try:
        anomaly_flags[col + '_anomaly'] = detect_arima_anomalies(df[col])
    except:
        anomaly_flags[col + '_anomaly'] = False

# --- Step 5: Combine anomaly results ---
# If any feature flags it, mark the row as anomalous
anomaly_flags['is_anomaly'] = anomaly_flags.any(axis=1)

# Merge with original DataFrame
df['is_anomaly'] = anomaly_flags['is_anomaly']

# Save final result
df.to_csv("arima_combined_anomalies.csv")
print(" Anomaly detection complete. Results saved to 'arima_combined_anomalies.csv'")
```
